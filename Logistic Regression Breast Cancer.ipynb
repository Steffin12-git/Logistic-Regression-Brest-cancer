```python
data = pd.read_csv(r"D:\Data\ML\Linear regression\Breast Cancer Wisconsin (Diagnostic) Data Set.csv")
data.head()
```

# Data cleaning

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['diagnosis'] = data['diagnosis'].apply(lambda x: 1 if str(x).strip().upper() == "M" else 0)
```

```python
print(data['diagnosis'].unique())
```

```python
#data["diagnosis"] = data["diagnosis"].astype("category", copy= False)
data["diagnosis"].value_counts().plot(kind="bar")
```

# Normalize data

```python
#dividing variables to target and predictators
y = data["diagnosis"]
X = data.drop(["diagnosis"], axis =1)
```

```python
# since all values are of different units first wee need to make it same unit
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

# Spliting data to training and testing set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

# train and predict the model

```python
from sklearn.linear_model import LogisticRegression

# create lr model
lr = LogisticRegression()

#train the model
lr.fit(X_train,y_train)

#predict the target variable
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
lr.coef_
```

# Evaluation of the model

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy : {accuracy: .2f}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_prob = lr.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc = roc_auc_score(y_test, y_prob)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
